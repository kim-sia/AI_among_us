```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import re
import plotly.offline as pyo
import plotly.graph_objs as go

plt.rcParams['figure.figsize'] = [20, 5]

df1 = pd.read_csv('../final_data/naver_news_scraping_ranking_0120_0831.csv')
df2 = pd.read_csv('../final_data/Time.csv')
```

```python
df1.info()
```

```python
# 결측치는 앞 행의 값으로 채우겠다!
df2 = df2.fillna(method='ffill')

df2.info()
```

```python
# df1 조회수를 날짜로 그룹화, 평균 구하기
df1_group = df1.groupby(by=df1['date'], as_index=False).mean()
df1_group
```

```python
df1_counts = df1['view']
df1_counts.mean()

criteria = df1_counts.quantile(q=0.98, interpolation='nearest') #상위 2프로의 뷰 숫자 
criteria
```

```python
# 그룹화 한 데이터 정렬하기
df1_group_sort = df1_group.sort_values('view', ascending=False)
df1_group_sort
```

```python
# 뉴스 데이터 추출 해 보기
news1 = df1[df1['date'] == 20200710]
print(news1)
```

```python
# df2의 날짜 데이터 형식 바꿔주기
df2['re_date'] = df2.date

for i,j in zip(df2['date'],range(len(df2)+1)):
    parse = re.sub('[-=.#/?:$}]', '', i)
    df2['re_date'][j] = parse
df2
```

```python
# 각 날짜의 일일 확진자 수 구하기
df2['confirmed_day'] = 0
df2['confirmed_day'][0] = df2['confirmed'][0]

for i in range(1, len(df2)) :
    df2['confirmed_day'][i] = df2['confirmed'][i] - df2['confirmed'][i-1]
    
df2
```

```python
# df2의 각 날짜에 조회수 평균 값 컬럼 추가하기
df2['view_avg'] = df1_group['view']
df2
```

```python
# 두 그래프 한번에 보자!
# 함수화

x = df2['re_date']
y1 = df2['confirmed_day']
y2 = df2['view_avg']

def print_graph(x, y1, y2) :

    fig, ax1 = plt.subplots()
    ax2 = ax1.twinx()

    line1 = ax1.plot(x, y1, linestyle='-', label='confirmed_day')
    line2 = ax2.plot(x, y2, color='darkorange', linestyle='-', label='view_avg')

    ax1.set_xticklabels(x, rotation=45)
    ax1.set_ylabel('number')
    ax2.set_ylabel('views')
    
#     ax1.set_ylim(0, 850)
#     ax2.set_ylim(0, 300000)
    lines = line1 + line2
    labels = [l.get_label() for l in lines]
    plt.legend(lines, labels)
    plt.show()
    
print_graph(x, y1, y2)
```

```python
# 랭킹 뉴스의 평균 조회수 순위가 30위까지 날짜 출력해보자

# 상위 30개만 추출하기
df1_group_rank = df1_group_sort[:30]

# 날짜로 정렬하기
df1_group_rank_sort_date = df1_group_rank.sort_values('date', ascending=True)
df1_group_rank_sort_date['date'].values
```

```python
# 상위 랭크 된 날짜 데이터 판별을 위해 df2에 새로운 컬럼 추가
df2['flag'] = 0
flag_list = df1_group_rank_sort_date['date'].values

for i in range(len(df2)) :
    if int(df2['re_date'][i]) in flag_list :
        df2['flag'][i] = 1
        
df2
```

```python
# 상위 30 데이터 그래프 그리기

df3 = df2.loc[:,['confirmed', 're_date','confirmed_day','view_avg','flag']]
def confirmed_num(n):
    for i in range(len(df3)-6):
        if df3['re_date'][i] == n:
#           print(confirmed.confirmed[i+6]-confirmed.confirmed[i])
            return df3['confirmed'][i+6] - df3['confirmed'][i]
            break
            return None
        else:
            continue
            
change = []
for i in df3['re_date'] :
    change.append(confirmed_num(i))
    
df3['change'] = change
            
x = df3['re_date'].loc[df2['flag']==1]
y1 = df3['change'].loc[df2['flag']==1]
y2 = df3['view_avg'].loc[df2['flag']==1]

print_graph(x, y1, y2)
```

```python
# df2에 날짜 구분을 위한 컬럼 생성
df2['day'] = 0
for i in range(len(df2)) :
    df2['day'][i] = df2['re_date'][i][6:8]
df2
```

```python
# 날짜에 따른 데이터 출력을 위한 데이터 프레임
confirmed = df2.loc[:, ['re_date','confirmed', 'confirmed_day', 'view_avg']]
confirmed

# 뉴스에 따른 확진자 숫자 양상 변화 함수
def confirmed_num(n):
    for i in range(len(confirmed)-6):
        if confirmed['re_date'][i] == n:
#           print(confirmed.confirmed[i+6]-confirmed.confirmed[i])
            return confirmed['confirmed'][i+6] - confirmed['confirmed'][i]
            break
            return None
        else:
            continue
            
# confirmed_num('20200606')
```

```python
change = []
for i in confirmed['re_date'] :
    change.append(confirmed_num(i))

df2['change'] = change
df2
```
